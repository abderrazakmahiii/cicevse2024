```python
import os
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Function to download a single file
def download_file(url, save_path):
    response = requests.get(url)
    with open(save_path, 'wb') as f:
        f.write(response.content)
    print(f"Downloaded {os.path.basename(save_path)}")

# Function to download CSV files from a directory URL
def download_csv_from_directory(base_url, save_folder):
    # Create directory to save files
    os.makedirs(save_folder, exist_ok=True)

    # Send a request to the URL
    response = requests.get(base_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Download all CSV files found in the directory
    for link in soup.find_all('a'):
        href = link.get('href')
        if href.endswith('.csv'):
            file_url = base_url + href
            save_path = os.path.join(save_folder, href)
            download_file(file_url, save_path)

# URLs for Network Traffic EVSE-A and EVSE-B directories
base_urls = [
    "http://205.174.165.80/IOTDataset/CICEVSE2024%20Dataset/Dataset/Network%20Traffic/EVSE-A/csv/",
    "http://205.174.165.80/IOTDataset/CICEVSE2024%20Dataset/Dataset/Network%20Traffic/EVSE-B/csv/"
]

# Download CSV files from both EVSE-A and EVSE-B directories
for base_url in base_urls:
    print(f"Downloading files from {base_url}...")
    download_csv_from_directory(base_url, "Dataset")

# Load and merge CSV files
csv_folder = "Dataset/"
csv_files = [os.path.join(csv_folder, f) for f in os.listdir(csv_folder) if f.endswith(".csv")]

# Load all CSV files into a dictionary of DataFrames
csv_dataframes = {os.path.basename(file): pd.read_csv(file) for file in csv_files}

# Add the 'label_mul' column and update the DataFrames
updated_dataframes = {
    file_name: df.assign(label_mul=" ".join(file_name.split(".")[0].split("-")[2:]))
    for file_name, df in csv_dataframes.items()
}

# Merge all updated DataFrames into a single DataFrame
merged_df = pd.concat(updated_dataframes.values(), ignore_index=True)

# Save the merged DataFrame to a new CSV file
merged_output_path = "./Network_Traffic.csv"
merged_df.to_csv(merged_output_path, index=False)

print(f"All files merged and saved to {merged_output_path}")
```
